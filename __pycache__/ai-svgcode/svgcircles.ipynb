```python
# Settings
square_size = 100
spacing = 10
text = list("Happy|Mother's|Day!")  # Using | as row break
rows = [list(row) for row in "Happy|Mother's|Day!".split("|")]
canvas_width = 800
canvas_height = 600
start_y = 100

colors = ["#e53935", "#1e88e5", "#43a047", "#fbc02d"]  # Swiss palette

svg_header = f'''<?xml version="1.0" encoding="UTF-8"?>
<svg width="{canvas_width}" height="{canvas_height}" xmlns="http://www.w3.org/2000/svg">
<style>
  .square-text {{
    font-family: Helvetica, Arial, sans-serif;
    font-size: 36px;
    fill: black;
    text-anchor: middle;
    dominant-baseline: central;
  }}
</style>
'''

svg_content = ""

for row_index, row in enumerate(rows):
    num_squares = len(row)
    total_width = num_squares * square_size + (num_squares - 1) * spacing
    start_x = (canvas_width - total_width) / 2
    y = start_y + row_index * (square_size + spacing)
    
    for col_index, char in enumerate(row):
        x = start_x + col_index * (square_size + spacing)
        color = colors[row_index % len(colors)]
        
        # Square
        svg_content += f'''
  <rect x="{x}" y="{y}" width="{square_size}" height="{square_size}" 
        fill="{color}" opacity="0.5" rx="10" ry="10"/>'''

        # Text
        svg_content += f'''
  <text x="{x + square_size / 2}" y="{y + square_size / 2}" class="square-text" opacity="1">{char}</text>'''

svg_footer = "\n</svg>"

# Save SVG
with open("mothers_day_swiss_grid.svg", "w") as f:
    f.write(svg_header + svg_content + svg_footer)

print("SVG saved as 'mothers_day_swiss_grid.svg'")
```
