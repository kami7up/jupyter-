```python
import IPython.display as display
import xml.etree.ElementTree as ET

# Define SVG parameters
dpi = 96  # dots per inch
inch = dpi  # 1 inch in pixels

# Circle details: (radius in inches, color)
circles = [
    (1, '#8B0000'),       # deep red
    (1.5, '#00008B'),     # dark blue
    (2, '#006400')        # dark green
]

# Determine SVG size (make sure all circles fit)
max_radius = max(r for r, _ in circles)
svg_size = int(2 * max_radius * inch)

# Create SVG root
svg = ET.Element('svg', width=str(svg_size), height=str(svg_size),
                 xmlns="http://www.w3.org/2000/svg")

# Center coordinates
cx = cy = svg_size / 2

# Add circles
for radius_inch, color in circles:
    radius_px = radius_inch * inch
    ET.SubElement(svg, 'circle', cx=str(cx), cy=str(cy),
                  r=str(radius_px), fill=color, opacity='0.7')

# Convert to string
svg_str = ET.tostring(svg).decode()

# Display SVG in notebook
display.display(display.SVG(svg_str))
```

```python
import svgwrite
from math import pi

# Conversion: inches to pixels (assuming 96 dpi)
inch = 96
radius1 = 1 * inch  # 1 inch
radius2 = 1.5 * inch  # 1.5 inches
radius3 = 2 * inch  # 2 inches

# Center position for all circles
center = (150, 150)

# Create SVG drawing
dwg = svgwrite.Drawing('overlapping_circles.svg', profile='tiny', size=('300px', '300px'))

# Draw circles
dwg.add(dwg.circle(center=center, r=radius3, fill='darkgreen', opacity=0.7))
dwg.add(dwg.circle(center=center, r=radius2, fill='darkblue', opacity=0.7))
dwg.add(dwg.circle(center=center, r=radius1, fill='darkred', opacity=0.7))

# Function to add curved (arc) text
def add_arc_text(dwg, text, radius, start_angle, color):
    # Center of the circle
    cx, cy = center
    # Approximate text placement along the arc
    # We'll use a simple method to split the text into characters and place along the arc
    total_angle = pi / 2  # 90 degrees on top semi-circle
    # Calculate angle per character
    angle_per_char = total_angle / len(text)
    start = start_angle

    for i, ch in enumerate(text):
        angle = start + i * angle_per_char
        x = cx + radius * -1 * (svgwrite.utils.cos(angle))
        y = cy + radius * -1 * (svgwrite.utils.sin(angle))
        # Rotate each character to align with the curve
        dwg.add(dwg.text(ch, insert=(x, y),
                         fill=color,
                         font_size='20px',
                         font_family='Arial',
                         dominant_baseline='middle',
                         text_anchor='middle',
                         transform=f'rotate({angle * 180 / pi} {x} {y})'))

# Add "Happy" on top of outer circle in curved style (dark green)
add_arc_text(dwg, "Happy", radius3 - 20, pi, 'darkgreen')  # pi radians (180 degrees to start at left)

# Add "Mother's" in the middle circle at top, center aligned, in curved style (deep red)
add_arc_text(dwg, "Mother's", radius2 - 20, pi/2, 'darkred')  # pi/2 radians (90 degrees)

# Add "Day" in the inner circle, straight and centered
dwg.add(dwg.text("Day", insert=center,
                 fill='darkblue',
                 font_size='24px',
                 font_family='Arial',
                 text_anchor='middle',
                 dominant_baseline='middle'))

# Save or display the SVG
dwg.save()

# Display the SVG inline in notebook
from IPython.display import SVG, display
display(SVG('overlapping_circles.svg'))
```

```python
from IPython.display import SVG, display
import math

# Function to convert inches to pixels (1 inch = 96 pixels)
def inch_to_px(inches):
    return inches * 96

# Define radii in pixels
r_outer = inch_to_px(2) / 2        # 2 inches
r_middle = inch_to_px(1.5) / 2     # 1.5 inches
r_inner = inch_to_px(1) / 2        # 1 inch

# Canvas size
canvas_size = 250  # pixels
cx = cy = canvas_size // 2  # center

# Create SVG path for curved text (arc on top of each circle)
def arc_path(radius):
    start_x = cx - radius
    end_x = cx + radius
    return f"M {start_x},{cy} A {radius},{radius} 0 0,1 {end_x},{cy}"

# Create the SVG content
svg_content = f'''
<svg width="{canvas_size}" height="{canvas_size}" xmlns="http://www.w3.org/2000/svg">
  <!-- Outer Circle -->
  <circle cx="{cx}" cy="{cy}" r="{r_outer}" fill="darkred" />
  <path id="path1" d="{arc_path(r_outer - 10)}" fill="none" />
  <text fill="darkgreen" font-size="24" font-family="Arial" text-anchor="middle">
    <textPath href="#path1" startOffset="50%">Happy</textPath>
  </text>

  <!-- Middle Circle -->
  <circle cx="{cx}" cy="{cy}" r="{r_middle}" fill="darkblue" />
  <path id="path2" d="{arc_path(r_middle - 10)}" fill="none" />
  <text fill="darkred" font-size="20" font-family="Arial" text-anchor="middle">
    <textPath href="#path2" startOffset="50%">Mother's</textPath>
  </text>

  <!-- Inner Circle -->
  <circle cx="{cx}" cy="{cy}" r="{r_inner}" fill="darkgreen" />
  <text x="{cx}" y="{cy + 8}" fill="darkblue" font-size="18" font-family="Arial" text-anchor="middle">Day</text>
</svg>
'''

# Display the SVG
display(SVG(svg_content))
```
